```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score
from sklearn.multioutput import MultiOutputClassifier
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.neural_network import MLPClassifier
data1 = pd.read_csv('T1.csv')
data2 = pd.read_csv('T2.csv')
data3 = pd.concat([data1, data2], axis=0, ignore_index=True)
data = data3
data = data.sample(frac=1, random_state=42).reset_index(drop=True)

X = data.iloc[:, 1:]  
y = data.iloc[:, 0]  

# Ensure labels are in list format for multilabel binarizer A
y = y.apply(lambda val: [val] if isinstance(val, str) else val)

# Convert labels to binary format
mlb = MultiLabelBinarizer()
y_encoded = mlb.fit_transform(y)

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.3, random_state=42
)
mlp = MLPClassifier(
    hidden_layer_sizes=5,
    activation='relu',
    solver='adam',
    max_iter=40,
    random_state=42
)
multi_label_mlp = MultiOutputClassifier(mlp)
```

```python
multi_label_mlp.fit(X_train, y_train)
```

```python
# Predict and evaluate
y_pred = multi_label_mlp.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=mlb.classes_))
```
